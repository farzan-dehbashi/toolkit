Seaborn Tutorial

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

# load Data

```python
crash_df = sns.load_dataset('car_crashes')
```

#  distrubution plot

```python
sns.distplot(crash_df['not_distracted'], kde=True, bins=25)
```

# joint plot

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='hex') #kind = kde, reg, hex
```

```python
sns.kdeplot(crash_df['alcohol'])
```

```python
sns.pairplot(crash_df)
```

```python
sns.rugplot(crash_df['alcohol'])
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(4,30))
sns.jointplot(x='speeding',y='alcohol', data=crash_df, kind='reg')
```

```python
import matplotlib.pyplot as plt

plt.plot([1,23,2,4])
plt.ylabel('some numbers')

plt.show()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(4,30))
sns.set_context('paper', font_scale=1.4) #paper talk poster
sns.jointplot(x='speeding',y='alcohol', data=crash_df, kind='reg')
```

# heat map

```python
plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale=1.4)

crash_mx = crash_df.corr()  #makes a corelation matrix by deviding the collomns by each other
sns.heatmap(crash_mx, annot=True, cmap='Blues')
```
